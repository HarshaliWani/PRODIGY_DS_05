```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('US_Accidents_March23.csv')
```

```python
df.head(10)
```

```python
df.count()
```

```python
df = df.sample(frac=0.1, random_state=1)
```

```python
datadict=pd.DataFrame(df.dtypes)
datadict['MissingVal']=df.isnull().sum()
datadict['UniqueVal']=df.nunique()
datadict['Count']=df.count()
datadict
```

```python
# Ensure datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
df['End_Time'] = pd.to_datetime(df['End_Time'], errors='coerce')

# Check for any NaT values after conversion
print(df[['Start_Time', 'End_Time']].isnull().sum())
```

```python
df['Hour'] = df['Start_Time'].dt.hour
df['Day_of_Week'] = df['Start_Time'].dt.day_name()
df['Month'] = df['Start_Time'].dt.month
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Start_Lng', y='Start_Lat', alpha=0.5)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

## Weather analysis for count of accidents.
### The plot shows that weather does not play major role in affecting the rate of accidents. Accident count is very high in fair weather as well

```python
# Group by Weather Condition and count occurrences
weather_analysis = df.groupby('Weather_Condition').size().reset_index(name='Counts')

# Sort the data and get the top 10
top_weather_analysis = weather_analysis.sort_values(by='Counts', ascending=False).head(10)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=top_weather_analysis, x='Weather_Condition', y='Counts')
plt.xticks(rotation=45)
plt.title('Top 10 Accidents by Weather Condition')
plt.show()
```

## Comparision of Road condition with accident
### Most of the road situations like bump, give-away, round about and turning loop see minimal to no accidents.
### A few accidents take place at traffic signals or crossings and the number is very low.

```python
# List of boolean columns to plot
bool_columns = [
    'Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 
    'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 
    'Traffic_Signal', 'Turning_Loop'
]

# Number of subplots
num_columns = len(bool_columns)
fig, axes = plt.subplots(nrows=(num_columns + 2) // 3, ncols=3, figsize=(15, 4 * ((num_columns + 2) // 3)))
axes = axes.flatten()  # Flatten the 2D array of axes

for i, col in enumerate(bool_columns):
    # Group by the current boolean column and count occurrences
    analysis = df.groupby(col).size().reset_index(name='Counts')
    
    # Plotting
    sns.barplot(data=analysis, x=col, y='Counts', ax=axes[i])
    axes[i].set_title(f'Accidents by {col}')
    axes[i].set_xticks([0, 1])
    axes[i].set_xticklabels(['False', 'True'])
    
# Hide any empty subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

## Severity of accidents.

### Maximum accidents, around 600000, that happen were of severity 2
### Around 100000 accidents are of severity level 3

```python
# Severity Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Severity')
plt.title('Distribution of Accident Severity')
plt.xlabel('Severity Level')
plt.ylabel('Number of Accidents')
plt.show()
```

## Accidents by the time of the day
### Maximum accidents happen early in the morning and evening, during crowded hours. When most of the people are travelling for job or school.

```python
# Severity by Hour of Day
df['Hour'] = df['Start_Time'].dt.hour
hour_severity = df.groupby('Hour')['Severity'].value_counts().unstack().fillna(0)
hour_severity.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Accident Severity by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=0)
plt.show()
```

```python
# Select relevant columns for correlation analysis
correlation_data = df[['Severity', 'Temperature(F)', 'Humidity(%)', 'Visibility(mi)']]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap of Environmental Variables and Severity')
plt.show()

# Pairplot to visualize relationships
sns.pairplot(correlation_data, hue='Severity', palette='viridis')
plt.suptitle('Pairplot of Environmental Variables vs Severity', y=1.02)
plt.show()
```

## Wind speed and direction
### We can see that wind does not affect the probability of an accident, as the accident count is high even when the wind is calm.

```python
# Clean Wind_Direction (if necessary)
df['Wind_Direction'] = df['Wind_Direction'].fillna('Unknown')  # Fill missing values

# Grouping by Wind Speed and Severity
speed_severity_analysis = df.groupby(['Wind_Speed(mph)', 'Severity']).size().unstack().fillna(0)

# Selecting the first 15 entries
top_speed_severity_analysis = speed_severity_analysis.head(20)

# Plotting
plt.figure(figsize=(12, 6))
top_speed_severity_analysis.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Accident Severity by Wind Speed')
plt.xlabel('Wind Speed (mph)')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

```python
# Analyzing Wind Direction's influence
direction_severity_analysis = df.groupby(['Wind_Direction', 'Severity']).size().unstack().fillna(0)

# Plotting for Wind Direction
plt.figure(figsize=(12, 6))
direction_severity_analysis.plot(kind='bar', stacked=True, colormap='plasma', figsize=(12, 6))
plt.title('Accident Severity by Wind Direction')
plt.xlabel('Wind Direction')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

## Cities with most accidents
### Miami has the maximum number of accidents.
### Dallas and Los Angeles has a comparatively high percentage of level 3 severity accidents.

```python
# Group by City, State, and Severity (or any other relevant column)
city_analysis = df.groupby(['City', 'Severity']).size().unstack().fillna(0)

# Sort the total accidents in descending order
city_analysis['Total'] = city_analysis.sum(axis=1)
top_cities = city_analysis.sort_values(by='Total', ascending=False).head(10)

# Plotting
plt.figure(figsize=(12, 6))
top_cities.drop(columns='Total').plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Top 10 Cities by Accident Severity')
plt.xlabel('City')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

```python
# Similarly for States or Counties
state_analysis = df.groupby(['State', 'Severity']).size().unstack().fillna(0)
state_analysis['Total'] = state_analysis.sum(axis=1)
top_states = state_analysis.sort_values(by='Total', ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_states.drop(columns='Total').plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Top 10 States by Accident Severity')
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```
